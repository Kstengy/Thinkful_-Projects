```python
import numpy as np
import matplotlib.pyplot as plt
import statistics
%matplotlib inline
```

```python
### Picking six diff distributions to practice on
# Binomial, Exponential, Gamma, Multinomial, Poisson and Wald
```

```python
#Binomial Distribution
#numpy.random.binomial(n, p, size=None)
Binomial = np.random.binomial(100, .25, 1000)
plt.hist((Binomial), color = 'g')
plt.axvline(Binomial.mean(), color = 'r', linestyle = 'solid', linewidth = 2)
plt.axvline(Binomial.mean() + Binomial.std(), color = 'r', linestyle = 'dashed', linewidth = 2)
plt.axvline(Binomial.mean() - Binomial.std(), color = 'r', linestyle = 'dashed', linewidth = 2)
plt.show()
```

```python
#Exponential Distribution
#numpy.random.exponential(scale=1.0, size=None)
Exponential = np.random.exponential(.75, 1000)
plt.hist(Exponential)
plt.axvline(Exponential.mean(), color = 'r', linestyle = 'solid', linewidth = 2)
plt.axvline(Exponential.mean() + Exponential.std(), color = 'r', linestyle = 'dashed', linewidth = 2)
plt.axvline(Exponential.mean() - Exponential.std(), color = 'r', linestyle = 'dashed', linewidth = 2)
plt.show()
```

```python
# Building a QQ plot to test the distribution above
Binomial.sort()
Exponential.sort()
plt.scatter(Binomial, Exponential)
x = [0, 40]
y = [0, 5]
plt.plot(x,y)
plt.show()
```

```python
#Gamma Distribution
#numpy.random.gamma(shape, scale=1.0, size=None)
Gamma = np.random.gamma(5, 1, 1000)
plt.hist((Gamma), color = 'r')
plt.axvline(Gamma.mean(), color = 'b', linestyle = 'solid', linewidth = 2)
plt.axvline(Gamma.mean() + Gamma.std(), color = 'b', linestyle = 'dashed', linewidth = 2)
plt.axvline(Gamma.mean() - Gamma.std(), color = 'b', linestyle = 'dashed', linewidth = 2)
plt.show()
```

```python
#Multinomial Distribution
#numpy.random.multinomial(n, pvals, size=None)
MultiNom = np.random.multinomial(20, [1/6]*6)
plt.axvline(MultiNom.mean(), color = 'black', linestyle = 'solid', linewidth = 2)
plt.axvline(MultiNom.mean() + MultiNom.std(), color = 'black', linestyle = 'dashed', linewidth = 2)
plt.axvline(MultiNom.mean() - MultiNom.std(), color = 'black', linestyle = 'dashed', linewidth = 2)
x = [1,2,3,4,5,6]
y = MultiNom
print(MultiNom)
plt.bar(x,y)
plt.show()
```

```python
#Poisson Distribution
# numpy.random.poisson(lam=1.0, size=None)
Poisson = np.random.poisson(5, 1000)
plt.hist(Poisson, color = 'purple')
plt.axvline(Poisson.mean(), color = 'y', linestyle = 'solid', linewidth = 2)
plt.axvline(Poisson.mean() + Poisson.std(), color = 'y', linestyle = 'dashed', linewidth = 2)
plt.axvline(Poisson.mean() - Poisson.std(), color = 'y', linestyle = 'dashed', linewidth = 2)
plt.show()
```

```python
# Wald distribution
# numpy.random.wald(mean, scale, size=None)
Wald = np.random.wald(50, 50, 1000)
plt.hist(Wald, color = 'k')
plt.axvline(Wald.mean(), color = 'c', linestyle = 'solid', linewidth = 2)
plt.axvline(Wald.mean() + Wald.std(), color = 'c', linestyle = 'dashed', linewidth = 2)
plt.axvline(Wald.mean() - Wald.std(), color = 'c', linestyle = 'dashed', linewidth = 2)
plt.show()
```

```python
Exponential.sort()
Wald.sort()
plt.scatter(Exponential, Wald)
plt.show()
```
